```python
import os, time

print("\033[1;35m Let's Play Animal Sounds!")
print("\033[1;37m________________________")
print("")

time.sleep(3.5)
os.system("clear")

exit = "no"
while exit == "no":
  animalSound = input("""\033[5;34m What animal do you want to hear?
    Please choose from (cat, dog, cow, wolf, duck, sheep, bird, snack, bee, chicken) :  """
                      )
  print("")
  if animalSound == "cow":
    print("\033[1;37m🐮: MOOOW")
    print("")
  elif animalSound == "cat":
    print("\033[1;37m🐱: MEEEOW")
    print("")
  elif animalSound == "bird":
    print("\033[1;37m🐦: SewSewSew")
    print("")
  elif animalSound == "sheep":
    print("\033[1;37m🐏: MAAAAH")
    print("")
  elif animalSound == "dog":
    print("\033[1;37m🐶: HawHaw")
    print("")
  elif animalSound == "chicken":
    print("\033[1;37m🐔: KaaaaaaK")
    print("")
  elif animalSound == "bee":
    print("\033[1;37m🐝: Zzzzzzz")
    print("")
  elif animalSound == "snack":
    print("\033[1;37m🐍: tssstsss")
    print("")
  elif animalSound == "duck":
    print("\033[1;37m🦆: QuackQuack")
    print("")
  elif animalSound == "wolf":
    print("\033[1;37m🐺: WOOOOO")
    print("")
  else:
    print("I don't know this animal please try again! ")
    print("")
  exit = input("\033[1;34m do you wanna exit? ")
  print("")
print("Bye!")
```
